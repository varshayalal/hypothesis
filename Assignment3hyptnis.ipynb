Null Hypothesis = u1 = u2

Alternative Hypothesis = u1 not equal to u2

Significance = 5%

alpha = 0.05

as it is two tailed then

alpha = 0.05/2 = 0.025

```python
data.isnull().sum()
```

```python
data.shape
```

```python
unitA = pd.Series(data.iloc[:,0])
```

```python
unitA
```

```python
unitB = pd.Series(data.iloc[:,1])
```

```python
unitB
```

```python
stats.ttest_ind(unitA, unitB)
```

```python
pvalue= 0.472
alpha= 0.025
```

```python
# now compare the value of p value and alpha

if pvalue < alpha:
  print('we reject null hypothesis')

else:

  print('we fail to reject null hypothesis')
```
